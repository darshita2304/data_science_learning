```python
#### final recursion methods.......
def split_cols_by_none(df1, df2=None):

    for col in df1.columns:
        set_ = set(df1[col])

        if (len(set_) == 1) and (list(set_)[0] == None):
            break

    if (col == df1.columns[len(df1.columns)-1]):  # ching last column in main
        return (df2, df1)

    df2 = df1.loc[:, 0:col-1]
    df1 = df1.loc[:, col+1:]
    
    df1 = split_cols_by_none(df1)
    
    return (df2, df1)
```

```python
def split_rows_by_none(df1, df2=None):
    
    for row in df1.index:
        set_ = set(df1.loc[row])
        # print(set_)
        if (len(set_) == 1) and (list(set_)[0] == None):
            break
    
    if (row == df1.index[len(df1.index)-1]):  # ching last column in main
        return (df2, df1)
    
    df2 = df1.loc[0:row-1, :]
    df1 = df1.loc[row + 1:, :]
    
    df1 = split_rows_by_none(df1)

    return (df2, df1)
```

```python
def flatten_tuple(nested_tuple):
    flattened_list = []
    for item in nested_tuple:
        if isinstance(item, tuple):
            flattened_list.extend(flatten_tuple(item))
        else:
            flattened_list.append(item)
    return tuple(flattened_list)


# # Example usage
# nested_tuple = ((1, 2, (3, 4)), (5, (6, 7)), 8, (9,))
# flattened_tuple = flatten_tuple(dd)
# print("Flattened tuple:", flattened_tuple)
```
