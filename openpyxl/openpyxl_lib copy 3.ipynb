```python
def get_rows_till_none(ls):
    
    if len(ls) == 0:
        return
    list1 = []
    list2 = []
    for item in ls:
        if item != None:
            list1.append(item)
        else:
            break
    
    list2 = ls[len(list1)+1:]
    return list1, get_rows_till_none(list2)
```

```python
def find_none_values(data, row=0, col=0):
    # Base case: Check if row index is out of range
    if row >= len(data):
        return

    # Check if current value is None
    if data[row][col] is None:
        print("None value found at row:", row, "column:", col)

    # Move to the next column
    next_col = col + 1

    # Check if the column index is out of range
    if next_col >= len(data[row]):
        # Move to the next row
        find_none_values(data, row + 1, 0)
    else:
        # Recursively explore horizontally
        find_none_values(data, row, next_col)


# Sample data
data = [
    ('Col1', 'Col2', 'Col3', None, 'Col11', 'Col12', None, 'Col111', 'Col122',
     None, 'Col44', 'Col55', 'Col66', None, 'Col44', 'Col55', 'Col66'),
    (100, 100, 100, None, 100, 100, None, 100, 100,
     None, 100, 100, 100, None, 100, 100, 100),
    # Add more rows here...
]

# Call the recursive function
find_none_values(rows)
```

```python
def find_none_values_rowwise(data, row=0, col=0):
    
    if row >= len(data):  #Check if row index is out of range
        return row

    # Check if current value is None
    if data[row][col] is None:
        return row
    else:
        return find_none_values_rowwise(data, row + 1, col)
```

```python
def find_none_values_colwise(data, row=0, col=0):

    if col >= len(data[0]):  # Check if col index is out of range
        return col

    # Check if current value is None
    # print(row,col)
    if data[row][col] is None:
        return col
    else:
        return find_none_values_colwise(data, row , col + 1)
```
