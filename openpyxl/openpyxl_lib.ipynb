```python
def get_rows(sheet):
    row_list = list()
    for row in sheet.iter_rows(values_only=True):
        row_list.append(row)
        
    return row_list
```

```python
rows = get_rows(main_sheet)
```

```python
def split_list_by_value(lst, split_value):
    
    sublists = []
    sublist = []

    for item in lst:
        
        if item != split_value:
            sublist.append(item)
        else:
            sublists.append(sublist)
            sublist = []
    # Append the remaining sublist after the last split value
    if(len(sublist)>0):
        sublists.append(sublist)

    return sublists


# Example usage:
my_list = [1, 2, 3, 0, 4, 5, 0, 6, 7,8,0]
split_value = 0
result = split_list_by_value(my_list, split_value)
print(result)
```

```python
def split_list_by_value_null(lst):
    split_value = None
    sublists = []
    sublist = []

    for item in lst:

        if item != split_value:
            sublist.append(item)
        else:
            sublists.append(sublist)
            sublist = []
    # Append the remaining sublist after the last split value
    if (len(sublist) > 0):
        sublists.append(sublist)

    return sublists


# Example usage:
my_list = [1, 2, 3, 0, 4, 5, 0, 6, 7, 8, 0]
split_value = 0
result = split_list_by_value(my_list, split_value)
print(result)
```
